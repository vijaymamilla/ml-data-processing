```python
df = bpropery_df.dropna()
```

```python
def convert_amenities_dict(amenities):
    #amenities: str = "Flooring##Balcony or Terrace##View##Parking Spaces##: 1##Lobby in Building##Flooring##Elevators in Building##: 1##Electricity Backup##Balcony or Terrace##Double Glazed Windows##View##Parking "
    if amenities is not None :
     
        amenities_list = amenities.replace("##:",":").split("##")
        amenities_dict = {}
        for amenity in amenities_list :
            if ':' in amenity:
                current_amenity = amenity.split(":")
                amenities_dict[current_amenity[0]] = current_amenity[1]
            else:
                amenities_dict[amenity] = True

        return amenities_dict
    
    else:
        return None
        
```

```python
df['amenities'] = df['amenities'].apply(convert_amenities_dict)
```

```python
df.to_csv('datasets/bproperty_pre_processing.csv',index= False)
```
